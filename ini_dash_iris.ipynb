```python
# Librerias
import dash
from dash import html, dcc
import plotly.express as px
from dash.dependencies import Input, Output
import pandas as pd
```

```python
# Cargar datos
df = px.data.iris()
df
```

```python
##Iniciar la aplicación

app = dash.Dash(__name__)

#Definicio de layput de la aplicacion

app.layout = html.Div([
    html.H1("Tablero de visualización de variables de características de orquideas"),
    html.P("Seleciciones las variables que desea relacionar para su respectivo análisis"),
    html.Label("Variable X: "),
    dcc.Dropdown(
        id="dropdown-x",
        options=[{"label":col, "value":col} for col in df.columns if col!='species'],
        value="sepal_length"
    ),
    html.Label("Variable-Y: "),
    dcc.Dropdown(
        id="dropdown-y",
        options=[{"label":col, "value":col} for col in df.columns if col!='species'],
        value="sepal_width"
    ),
    html.Div([
        dcc.Graph(id="scatter-plot"),
        dcc.Graph(id="histogram"),
        dcc.Graph(id="box-plot")
    ])

])
```

```python
#Configuracion de las graficas y metodos de cambio automatico

#Configuración de las graficas y metodos de cambio automatico

@app.callback(
    Output("scatter-plot","figure"),
    [Input("dropdown-x","value"),
     Input("dropdown-y","value")]
)

def update_graph(x_col,y_col):
  fig=px.scatter(df, x=x_col, y=y_col, color='species',title=f"Scatter-plot de {x_col} vs {y_col}")
  return fig

@app.callback(
    Output("histogram","figure"),
    [Input("dropdown-x","value")]
)

def update_graph(x_col):
  fig=px.histogram(df, x=x_col, color='species',title=f"Histograma de {x_col}")
  return fig

@app.callback(
    Output("box-plot","figure"),
    [Input("dropdown-x","value")]
)

def update_graph(x_col):
  fig=px.box(df, x='species', y=x_col, title=f"Box-plot de {x_col} vs especies")
  return fig
```

```python
#Ejecutar

if __name__ == '__main__':
    app.run(debug=True)
```
