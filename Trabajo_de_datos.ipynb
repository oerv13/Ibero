```python
# Cargue de Librerias
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Configuración para visualizaciones
%matplotlib inline
sns.set(style="whitegrid")
```

```python
# Ver las primeras filas del conjunto de datos
print(df.head())

# Información general de columnas, tipos de datos y valores nulos
print(df.info())

# Descripción estadística para columnas numéricas
print(df.describe())
```

```python
# Histograma para una columna numérica
plt.figure(figsize=(8, 6))
df['Players'].hist(bins=30, color='skyblue', edgecolor='black')
plt.title('Distribución de la Columna Players')
plt.xlabel('Valores')
plt.ylabel('Frecuencia')
plt.show()
```

```python
# Gráfico de densidad
sns.kdeplot(df['Players'], color='red', shade=True)
plt.title('Gráfico de Densidad')
```

```python
# Boxplot para entender la distribución de una columna categórica con respecto a una numérica
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Parameter', y='Players', palette='Set2')
plt.title('Diagrama de Cajas')
plt.xticks(rotation=45)
plt.show()
```

```python
# Pairplot para analizar relaciones entre múltiples variables
sns.pairplot(df, hue='Parameter', palette='viridis')
plt.show()
```

```python
# Usando Boxplot para detectar outliers
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, y='Players', color='orange')
plt.title('Boxplot para Detección de Outliers')
plt.show()
```

```python
# import pandas as pd
import numpy as np
from scipy import stats
```

```python
# Calculo de la Media
media = df['Players'].mean()
print("Media:", media)
```

```python
# Calculo de la Mediana
mediana = df['Players'].median()
print("Mediana:", mediana)
```

```python
# Calculo de la Moda
moda = stats.mode(df['Players']).mode[1]
print("Moda:", moda.mode[1])

# Calcula la moda utilizando la función mode() de statistics
#mode =  df['Players'].mode()
#print("Moda:", mode)
```

```python
#df.Players.describe()
stats.mode(df['Players'])[1]
```

```python
# Calculo de la Varianza
varianza = df['Players'].var()
print("Varianza:", varianza)
```

```python
# Calculo Desviación Estándar
desviacion_estandar = df['Players'].std()
print("Desviación Estándar:", desviacion_estandar)
```

```python
# Calculo Coeficiente de Variación
coef_variacion = (desviacion_estandar / media) * 100
print("Coeficiente de Variación (%):", coef_variacion)
```

```python
# Calculo de la Normalización Z (Z-score)
df['Z-score'] = (df['Players'] - media) / desviacion_estandar
print(df[['Players', 'Z-score']].head())
```
