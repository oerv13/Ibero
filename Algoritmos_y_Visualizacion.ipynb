Definicin archivo Nacha
Los archivos NACHA son el formato de archivo estandarizado que utilizan los bancos para iniciar lotes de pagos ACH . Estos archivos ayudan a los bancos a ejecutar grandes volmenes de pagos ACH a travs de la Cmara de Compensacin (TCH) y la Reserva Federal.
Cuando se realiza una transaccin, esta puede tener como resultado una ejecucin exitosa o un rechazo. En los casos en que la transaccin es rechazada, se generan cdigos de respuesta como R01, R02, entre otros que permiten identificar la causa especfica de la novedad o el motivo del rechazo. Estos cdigos son fundamentales para el anlisis y monitoreo de las operaciones. Adems, es importante considerar el tipo de transaccin como una variable clave en los procesos de anlisis de datos, ya que aporta contexto adicional para la interpretacin de los resultados.

```python
# Calculo de la MOda
moda = df.groupby('NAMEENTIDADORIGEN')['VALORTRANSAC'].agg(lambda x: x.mode().iloc[0] if not x.mode().empty else np.nan)
moda = moda.reset_index().rename(columns={'VALORTRANSAC': 'Moda'})
print(moda)
plt.figure(figsize=(10,6))
plt.bar(moda['NAMEENTIDADORIGEN'], moda['Moda'], color='purple')
plt.title('Moda de VALORTRANSAC por NAMEENTIDADORIGEN')
plt.xlabel('NAMEENTIDADORIGEN')
plt.ylabel('VALORTRANSAC')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Banco de Bogot tiene una barra alta con un valor de moda cercano a 1.1e9 (1.100.000.000).

Esto indica que una transaccin con ese valor especfico se repite muchas veces en los datos de esa entidad.

Las dems entidades (BBVA, Bancolombia, Citibank, Itau) no tienen una moda destacable (barra nula o muy baja).


```python
# Calcular varianza, desviacin estndar y coeficiente de variacin
stats = df.groupby('NAMEENTIDADORIGEN')['VALORTRANSAC'].agg([
    ('Varianza', 'var'),
    ('Desviacin estndar', 'std'),
    ('Media', 'mean')
])

# Agregar columna de coeficiente de variacin: (std / mean) * 100
stats['Coeficiente de variacin (%)'] = (stats['Desviacin estndar'] / stats['Media']) * 100

# Eliminar la columna de Media si solo quieres graficar los tres indicadores
stats = stats.drop(columns='Media')

# Reset index para poder graficar
stats = stats.reset_index()

# Grfica
plt.figure(figsize=(12,6))
x = np.arange(len(stats['NAMEENTIDADORIGEN']))

# Ancho de las barras
width = 0.25

# Barras
plt.bar(x - width, stats['Varianza'], width=width, label='Varianza', color='skyblue')
plt.bar(x, stats['Desviacin estndar'], width=width, label='Desviacin Estndar', color='orange')
plt.bar(x + width, stats['Coeficiente de variacin (%)'], width=width, label='Coeficiente de Variacin (%)', color='green')

# Ejes y ttulos
plt.xlabel('NAMEENTIDADORIGEN')
plt.ylabel('Valor')
plt.title('Varianza, Desviacin Estndar y Coeficiente de Variacin de VALORTRANSAC')
plt.xticks(x, stats['NAMEENTIDADORIGEN'], rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

Si una entidad tiene alta varianza y desviacin estndar, significa que sus transacciones varan mucho en monto.

Pero si su coeficiente de variacin es bajo, quiere decir que esa variacin es proporcionalmente baja respecto al valor medio. Es decir, la entidad tiene operaciones grandes, pero estables.

Por el contrario, si una entidad tiene un coeficiente de variacin alto, sus montos son ms impredecibles, aunque la media no sea tan alta.

Dado que el monto de BBVA esta muy alto, no se lograr ver la informacin de las otras variables, afectando un anlisis ms cuidadoso.

```python
# Agrupamos los datos
grouped = df.groupby(['IDENTIDADORIGEN', 'CODTRANSAC', 'CODRESP']).size().reset_index(name='Cuenta')

# Crear una columna combinada para el eje X
grouped['Entidad_Transac'] = grouped['IDENTIDADORIGEN'].astype(str) + ' - ' + grouped['CODTRANSAC'].astype(str)

# Grfico de barras
plt.figure(figsize=(16, 7))
sns.barplot(data=grouped, x='Entidad_Transac', y='Cuenta', hue='CODRESP', ci=None)

plt.title('Distribucin de CODRESP por IDENTIDADORIGEN y CODTRANSAC')
plt.xlabel('Entidad y Cdigo de Transaccin')
plt.ylabel('Cantidad')
plt.xticks(rotation=45, ha='right')
plt.legend(title='CODRESP', bbox_to_anchor=(1.02, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

Distribucin desigual entre entidades:

Algunas entidades (Banco de Bogot, Bancolombia, etc.) pueden tener muchas ms transacciones que otras.

Esto puede indicar volmenes operativos distintos entre las instituciones.
Prevalencia de ciertos CODTRANSAC:

Es probable que ciertos cdigos de transaccin como 22 y 32 sean los ms frecuentes.

Presencia de errores o respuestas particulares (CODRESP):

Si hay mltiples colores en una misma barra, indica que una misma combinacin de entidad y transaccin tiene distintos resultados.

Si una categora especfica de CODRESP predomina (por ejemplo, R04 o R35), puede ser un indicador de problemas frecuentes o rechazos en ciertas combinaciones.
