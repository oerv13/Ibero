Redeban, compañía especializada en medios de pago electrónicos, nació en 2000 tras la fusión de dicha compañía con Red Multicolor. Entre sus accionistas se encuentran bancos como Davivienda, Bancolombia, Banco Caja Social, Bbva, Banco de Bogotá, Scotiabank Colpatria, Banco GNB Sudameris e Itaú.

En la era digital, la administración y el análisis de grandes volúmenes de datos se han convertido en una necesidad esencial para las empresas del sector financiero y de tecnología. El Big Data permite almacenar, procesar y analizar información de manera eficiente, brindando a las organizaciones la capacidad de tomar decisiones basadas en datos en tiempo real. En este contexto, los archivos Nacham, que contienen una gran cantidad de transacciones en cada ciclo, requieren de un sistema de almacenamiento robusto y una base de datos estructurada para gestionar la información de manera eficiente.

El manejo adecuado de estos archivos permite obtener estadísticas clave sobre los tipos de transacción, rechazos y devoluciones por entidad financiera. Con esta información, las empresas pueden analizar patrones de errores y optimizar los procesos para reducir la cantidad de rechazos, mejorando la eficiencia operativa y la experiencia del usuario. La implementación de técnicas de Big Data en la gestión de archivos Nacham es un paso esencial hacia la automatización y optimización de las transacciones financieras.

Definición archivo Nacha
Los archivos NACHA son el formato de archivo estandarizado que utilizan los bancos para iniciar lotes de pagos ACH . Estos archivos ayudan a los bancos a ejecutar grandes volúmenes de pagos ACH a través de la Cámara de Compensación (TCH) y la Reserva Federal.
Cuando se realiza una transacción, esta puede tener como resultado una ejecución exitosa o un rechazo. En los casos en que la transacción es rechazada, se generan códigos de respuesta como R01, R02, entre otros que permiten identificar la causa específica de la novedad o el motivo del rechazo. Estos códigos son fundamentales para el análisis y monitoreo de las operaciones. Además, es importante considerar el tipo de transacción como una variable clave en los procesos de análisis de datos, ya que aporta contexto adicional para la interpretación de los resultados.

BBVA Colombia tiene una media extremadamente alta, mucho mayor que la mediana.

Otras entidades como Bancolombia, Citibank, e Itau tienen valores bajos tanto en media como en mediana, pero en algunos casos la media sigue siendo ligeramente mayor que la mediana, lo cual también sugiere posibles outliers (aunque mucho menores que en BBVA).


```python
# Calculo de la MOda
moda = df.groupby('NAMEENTIDADORIGEN')['VALORTRANSAC'].agg(lambda x: x.mode().iloc[0] if not x.mode().empty else np.nan)
moda = moda.reset_index().rename(columns={'VALORTRANSAC': 'Moda'})
print(moda)
plt.figure(figsize=(10,6))
plt.bar(moda['NAMEENTIDADORIGEN'], moda['Moda'], color='purple')
plt.title('Moda de VALORTRANSAC por NAMEENTIDADORIGEN')
plt.xlabel('NAMEENTIDADORIGEN')
plt.ylabel('VALORTRANSAC')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Banco de Bogotá tiene una barra alta con un valor de moda cercano a 1.1e9 (1.100.000.000).

Esto indica que una transacción con ese valor específico se repite muchas veces en los datos de esa entidad.

Las demás entidades (BBVA, Bancolombia, Citibank, Itau) no tienen una moda destacable (barra nula o muy baja).


```python
# Calcular varianza, desviación estándar y coeficiente de variación
stats = df.groupby('NAMEENTIDADORIGEN')['VALORTRANSAC'].agg([
    ('Varianza', 'var'),
    ('Desviación estándar', 'std'),
    ('Media', 'mean')
])

# Agregar columna de coeficiente de variación: (std / mean) * 100
stats['Coeficiente de variación (%)'] = (stats['Desviación estándar'] / stats['Media']) * 100

# Eliminar la columna de Media si solo quieres graficar los tres indicadores
stats = stats.drop(columns='Media')

# Reset index para poder graficar
stats = stats.reset_index()

# Gráfica
plt.figure(figsize=(12,6))
x = np.arange(len(stats['NAMEENTIDADORIGEN']))

# Ancho de las barras
width = 0.25

# Barras
plt.bar(x - width, stats['Varianza'], width=width, label='Varianza', color='skyblue')
plt.bar(x, stats['Desviación estándar'], width=width, label='Desviación Estándar', color='orange')
plt.bar(x + width, stats['Coeficiente de variación (%)'], width=width, label='Coeficiente de Variación (%)', color='green')

# Ejes y títulos
plt.xlabel('NAMEENTIDADORIGEN')
plt.ylabel('Valor')
plt.title('Varianza, Desviación Estándar y Coeficiente de Variación de VALORTRANSAC')
plt.xticks(x, stats['NAMEENTIDADORIGEN'], rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

Si una entidad tiene alta varianza y desviación estándar, significa que sus transacciones varían mucho en monto.

Pero si su coeficiente de variación es bajo, quiere decir que esa variación es proporcionalmente baja respecto al valor medio. Es decir, la entidad tiene operaciones grandes, pero estables.

Por el contrario, si una entidad tiene un coeficiente de variación alto, sus montos son más impredecibles, aunque la media no sea tan alta.

Dado que el monto de BBVA esta muy alto, no se lograr ver la información de las otras variables, afectando un análisis más cuidadoso.


```python
# Gráfico circular (pie): Respuestas CODRESP
plt.figure(figsize=(6,6))
df['CODRESP'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Distribución CODRESP')
plt.ylabel('')
plt.show()
```

R35 - Cuenta restringida (46.3%), Es el motivo de rechazo más frecuente. Indica que casi la mitad de las transacciones fallidas no se pudieron procesar porque la cuenta tenía restricciones.

R17 - Error de codificación (24.1%), El segundo más común. Se refiere a errores en el formato de la transacción o información mal digitada, lo cual requiere revisión manual. Esto podría apuntar a problemas en el origen del archivo o en la automatización del proceso.

R04 - No cuenta (21.1%), Indica que el número de cuenta no corresponde a ninguna cuenta activa. Esto sugiere errores en el ingreso del número de cuenta o que la cuenta fue cerrada sin actualizar los datos.

R02 - Cuenta cerrada (6.4%), Ocurre cuando el banco receptor ha cerrado la cuenta del destinatario. Puede ser por baja de cuenta o movimientos no autorizados previos.

Otros códigos (menor al 1.5% cada uno): Como R88, R16, R33, etc., son menos frecuentes, pero aún importantes, ya que indican situaciones específicas como cuentas congeladas, tipos de cuenta no válidos, o pagos parados.

Esto indica que transacciones con respuesta R04 tienden a involucrar montos mucho más altos que el resto.

```python
# Agrupamos los datos
grouped = df.groupby(['IDENTIDADORIGEN', 'CODTRANSAC', 'CODRESP']).size().reset_index(name='Cuenta')

# Crear una columna combinada para el eje X
grouped['Entidad_Transac'] = grouped['IDENTIDADORIGEN'].astype(str) + ' - ' + grouped['CODTRANSAC'].astype(str)

# Gráfico de barras
plt.figure(figsize=(16, 7))
sns.barplot(data=grouped, x='Entidad_Transac', y='Cuenta', hue='CODRESP', ci=None)

plt.title('Distribución de CODRESP por IDENTIDADORIGEN y CODTRANSAC')
plt.xlabel('Entidad y Código de Transacción')
plt.ylabel('Cantidad')
plt.xticks(rotation=45, ha='right')
plt.legend(title='CODRESP', bbox_to_anchor=(1.02, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

Distribución desigual entre entidades:

Algunas entidades (Banco de Bogotá, Bancolombia, etc.) pueden tener muchas más transacciones que otras.

Esto puede indicar volúmenes operativos distintos entre las instituciones.
Prevalencia de ciertos CODTRANSAC:

Es probable que ciertos códigos de transacción como 22 y 32 sean los más frecuentes.

Presencia de errores o respuestas particulares (CODRESP):

Si hay múltiples colores en una misma barra, indica que una misma combinación de entidad y transacción tiene distintos resultados.

Si una categoría específica de CODRESP predomina (por ejemplo, R04 o R35), puede ser un indicador de problemas frecuentes o rechazos en ciertas combinaciones.

Identificación de patrones de falla:

Por ejemplo, si Bancolombia - 22 tiene una gran cantidad de transacciones con

R04, se podría investigar qué está fallando con esa transacción específica en esa entidad.


```python
# Filtrar solo columnas numéricas
df_numeric = df.select_dtypes(include='number')

# Calcular la matriz de correlación
correlation_matrix = df_numeric.corr()

# Crear el heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', linewidths=0.5)

plt.title('Mapa de calor de correlación entre variables numéricas')
plt.tight_layout()
plt.show()
```

**Conclusiones**

Alta concentración de transacciones en pocos tipos (CODTRANSAC):
El 73.4% de las transacciones se agrupan en el código 22 y el 26% en el 32. Esto representa un alto grado de concentración operativa, lo que puede facilitar la gestión de procesos repetitivos, pero también aumenta la exposición a errores sistemáticos en esos tipos específicos.

Variabilidad operativa entre entidades financieras:
Entidades como Banco de Bogotá, Bancolombia e Itaú ejecutan una mayor diversidad de transacciones, lo que puede indicar un portafolio más amplio o una operación más compleja. En contraste, entidades como Citibank o BBVA Colombia muestran menor variedad o volumen de operaciones.

Presencia de valores atípicos en montos (VALORTRANSAC):
Se detectan outliers significativos, especialmente para transacciones con código de respuesta R04, con montos que superan los 5 × 10¹¹. Esto podría deberse a errores de carga, transacciones corporativas de gran volumen o incluso eventos fraudulentos.

Diferentes resultados en transacciones similares (CODRESP):
Para una misma entidad y tipo de transacción, se observan múltiples códigos de respuesta (CODRESP), reflejando un porcentaje no menor de transacciones fallidas o rechazadas. Los códigos R04 y R35 aparecen de forma frecuente, lo que puede indicar ineficiencias operativas o validaciones que requieren revisión.

Riesgo operativo y necesidad de control:
Las entidades con mayor volumen deben implementar mecanismos automatizados de monitoreo y control, especialmente en los procesos asociados a los CODTRANSAC más frecuentes y con mayores tasas de rechazo.

Adopción de soluciones basadas en Big Data y analítica avanzada:

Automatización del monitoreo de transacciones y respuestas (CODRESP):
Implementar sistemas de alertas automatizadas para respuestas críticas como R04 o R35, integradas con dashboards de operaciones y control de calidad.

Implementar validaciones preventivas en tiempo real:
Incorporar lógica de negocio en los puntos de origen para prevenir transacciones que no cumplen con reglas básicas antes de ser procesadas.

Desarrollar modelos predictivos para estimar probabilidad de error o rechazo:
