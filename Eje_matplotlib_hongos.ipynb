```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/mushrooms.csv')
df.head()
```

```python
#crear figura

fig, axs = plt.subplots(2,2, figsize=(10,10))
fig.suptitle("Análisis exploratorio datos de clasificación de hongos")

#Grafico de conteo clases
clases = df['class'].value_counts()
axs[0,0].bar(clases.index, clases.values)
axs[0,0].set_title("Distribución de clases")
axs[0,0].set_ylabel("Frecuencia")
axs[0,0].set_xlabel("Clase")
#agregar valores a la grafica
for i, v in enumerate(clases.values):
    axs[0,0].text(i, v, str(v), ha='center', va='bottom')

# Conteo del color del sombrero
color = df['cap-color'].value_counts()
axs[0,1].bar(color.index, color.values)
axs[0,1].set_title("Distribución del color del sombrero")
axs[0,1].set_ylabel("Frecuencia")
axs[0,1].set_xlabel("Color")
#agregar valores a la grafica
for i, v in enumerate(color.values):
    axs[0,1].text(i, v, str(v), ha='center', va='bottom')

#Conteo del olor
odor = df['odor'].value_counts()
axs[1,0].bar(odor.index, odor.values)
axs[1,0].set_title("Distribución del olor")
axs[1,0].set_ylabel("Frecuencia")
axs[1,0].set_xlabel("Olor")
#agregar valores a la grafica
for i, v in enumerate(odor.values):
    axs[1,0].text(i, v, str(v), ha='center', va='bottom')

#Conteo de tamaño de los anillos
ring = df['ring-number'].value_counts()
axs[1,1].bar(ring.index, ring.values)
axs[1,1].set_title("Distribución del tamaño de los anillos")
axs[1,1].set_ylabel("Frecuencia")
axs[1,1].set_xlabel("Tamaño de los anillos")
#agregar valores a la grafica
for i, v in enumerate(ring.values):
    axs[1,1].text(i, v, str(v), ha='center', va='bottom')
```

```python
#Grafica de pastel
habitat = df['habitat'].value_counts()
plt.pie(habitat.values, labels=habitat.index, autopct='%1.1f%%')
plt.title("Distribución de habitats")
plt.show()
```
