```python
# Lectura del archivo csv, separandolo con ;
url = 'ArcNachas1.csv'
df = pd.read_csv(url, sep=';', encoding='latin-1')
df.columns = df.columns.str.replace('"', '')
df.columns = df.columns.str.replace(';', ',')
df = df.apply(lambda x: x.str.replace(',', '.') if x.dtype == "object" else x)
df = df.apply(lambda x: x.str.replace(';', ',') if x.dtype == "object" else x)
df = df.apply(lambda x: x.str.replace('"', '') if x.dtype == "object" else x)
print(df.head())
```
