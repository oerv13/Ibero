```python
# Calcular la media de una columna específica
media_columna = df["VALORTRANSAC"].mean()
print("La media de VALORTRANSAC es:", media_columna)
```

```python
# Calcular la media multiple columna
df[["VALORTRANSAC", "CODTRANSAC"]].mean()
```

```python
# Calcular la mediana de una columna específica
mediana_columna = df["VALORTRANSAC"].median()
print("La mediana de VALORTRANSAC es:", mediana_columna)
```

```python
# Calcular la moda de una columna específica
moda_columna = df["VALORTRANSAC"].mode()

print("La moda de VALORTRANSAC es:", moda_columna[0])
```

```python
# Calcular la varianza de una columna específica
varianza_columna = df["VALORTRANSAC"].var()

print("La varianza de VALORTRANSAC es:", varianza_columna)
```

```python
# Calcular la desviación estándar de una columna específica
desviacion_columna = df["VALORTRANSAC"].std()

print("La desviación estándar de VALORTRANSAC es:", desviacion_columna)
```

```python
# Seleccionar la columna numérica a analizar
columna = "VALORTRANSAC"

# Calcular media y desviación estándar
media = df[columna].mean()
desviacion = df[columna].std()

# Calcular coeficiente de variación
cv = (desviacion / media) * 100

print(f"El coeficiente de variación de {columna} es: {cv:.2f}%")
```

```python
# Seleccionar la columna a normalizar
columna = "VALORTRANSAC"

# Aplicar la normalización Z-Score
df["z_score"] = (df[columna] - df[columna].mean()) / df[columna].std()

print(df.head())
```

```python
# Cargue de Librerias
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Configuración para visualizaciones
%matplotlib inline
sns.set(style="whitegrid")
```

```python
# Histograma para una columna numérica
plt.figure(figsize=(8, 6))
df['CODRESP'].hist(bins=30, color='skyblue', edgecolor='black')
plt.title('Distribución de la Columna Codigo de Respuesta')
plt.xlabel('Valores')
plt.ylabel('Frecuencia')
plt.show()
```
